```python
from datasets import load_dataset
from PIL import Image
import os

dataset = load_dataset("afaji/cvqa")
```

```python
# print languages in the CVQA dataset and their respective number of samples
# structure: DatasetDict({
#     test: Dataset({
#         features: ['image', 'ID', 'Subset', 'Question', 'Translated Question', 'Options', 'Translated Options', 'Label', 'Category', 'Image Type', 'Image Source', 'License'],
#         num_rows: 10374
#     })
# })

from collections import defaultdict

test_split = dataset["test"]
language_counts = defaultdict(int)
for sample in test_split:
    language = sample["Subset"]
    language_counts[language] += 1

# Print results
print("Languages and sample counts in CVQA test split:")
for lang, count in sorted(language_counts.items(), key=lambda x: x[1], reverse=True):
    print(f"- {lang}: {count} samples")

print(f"\nTotal test samples: {sum(language_counts.values())}")
```

```python
languages = [
    'Amharic',
    'Bengali',
    'Chinese',
    'Egyptian_Arabic',
    'Filipino',
    'Hindi',
    'Japanese',
    'Korean',
    'Norwegian',
    'Portuguese',
    'Romanian',
    'Russian',
    'Spanish',
    'Swahili',
    'Tamil',
    'Telugu',
    'Urdu'
]

for row in dataset["test"]:
    language = eval(row["Subset"])[0]
    
    if language in languages:
        output_dir = f"{language}_images"
        os.makedirs(output_dir, exist_ok=True)
        image_path = os.path.join(output_dir, f'{row["ID"]}.jpg')
        
        if not os.path.exists(image_path):
            image: Image.Image = row['image']
            image.save(image_path)
            print(f"Saved {row['ID']} to {image_path}")
        else:
            print(f"Image {row['ID']} already exists, skipping.")
```
